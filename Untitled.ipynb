```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('full.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
#Fill null values in the Age column
age = data['Age']
age_wiki = data['Age_wiki']
age.fillna(age_wiki, inplace = True)

age.fillna(age.mean(), inplace = True)
age.isnull().sum()
```

```python
#Drop unnecessary columns
data.drop('WikiId', inplace = True, axis=1)
data.drop('Age_wiki', inplace = True, axis=1)
data.drop('Name_wiki', inplace = True, axis=1)
data.drop('Lifeboat', inplace = True, axis=1)
```

```python
# Fill null values in the Fare column
fare = data['Fare']
fare_mean = fare.mean()
fare.fillna(fare_mean, inplace = True)
```

```python
# Fill null values in the Embarked column
data['Embarked'] = data['Embarked'].fillna(data['Embarked'].mode()[0])
```

```python
# Fill null values in the Hometown column
data['Hometown'] = data['Hometown'].fillna(data['Hometown'].mode()[0])
```

```python
# Fill null values in the Boarded column
data['Boarded'] = data['Boarded'].fillna(data['Boarded'].mode()[0])
```

```python
# Fill null values in the Destination column
data['Destination'] = data['Destination'].fillna(data['Destination'].mode()[0])
```

```python
# Fill null values in the Class column
data['Class'] = data['Class'].fillna(data['Class'].mode()[0])
```

```python
data.isnull().sum()
```

```python
#Drop the columns that are missing values more than 70%
threshold = 0.7

data = data[data.columns[data.isnull().mean()<threshold]]
```

```python
data.columns
```

```python
data.sample(5)
```

```python
data['Survived'] = data['Survived'].fillna(data['Survived'].median())
```

```python
data.isnull().sum()
```

### Feature split

```python
data['HomeCountry'] = data.Hometown.str.split(" ").map(lambda x : x[-1])
#data['Hometown'] = data.Hometown.str.split(" ").map(lambda x : x[0])
```

```python
data.sample(8)
```

```python
data['Destination_Country'] = data.Destination.str.split(" ").map(lambda x : x[-1])
```

```python
data.sample(10)
```

### Label Encoding

```python
from sklearn import preprocessing
```

```python
label_encoder = preprocessing.LabelEncoder()
data['Sex'] = label_encoder.fit_transform(data['Sex'])
```
